# Gráficos

> A partir do que foi visto em sala, faça os seguintes gráficos:

- Distribuição da quantidade de sobreviventes.
- Sexo em relação a quantidade de sobreviventes.
- Boxplot para destribuição de sobreviventes por idade.
- Histograma de idade.

> Faça uma pequena análise para cada gráfico com as informações relevantes obtidas através das visualizações. E caso encontre, aponte possíveis outliers/anomalias.

```python
plt.figure(figsize=(8,6))
sns.countplot(x="survived",data=df)
plt.title("Distribuição da quantidade de sobreviventes")
plt.xlabel("Survived")
plt.ylabel("Contagem/Quantidade")
plt.show()
```

Conseguimos perceber que a quantidade de sobreviventes é menor comparada a quantidade de nao sobreviventes


```python
#Sexo em relação a quantidade de sobreviventes.
plt.figure(figsize=(8,6))
sns.countplot(x="sex",hue="survived",data=df)
plt.title("Sexo em relação a quantidade de sobreviventes")
plt.xlabel("Sex")
plt.ylabel("Contagem/Quantidade")
plt.show()
```

Percebemos que no caso de nao sobreviveu -> A maioria que nao sobreviveu foi homem
Ja no caso de sobreviveu -> As mulheres foram as que mais sobreviveram

```python
#Boxplot para destribuição de sobreviventes por idade.
sns.boxplot(y="age",x = "survived",data = df,hue = "survived")
plt.title("Distribuicao de sobreviventes por idade")
plt.xlabel("Survived")
plt.ylabel("Age/idade")
plt.legend(title='Nao sobreviveu : 0 \n Sobreviveu : 1')
plt.show()
```

Percebemos que em ambos a maioria dos casos é parecido, entre 20 a 40 anos mas no "nao sobreviveu" temos mais casos extremos com pessoas bem mais velhas ou muito novas----- Ja no sobreviveu temos casos extremos de pessoas mais velhas

```python
#Histograma de idade.
sns.histplot(df["age"],kde = True)
plt.ylabel("Frequencia")
plt.xlabel("Idade/age")
plt.title("Distribuicao de idade")
plt.show()
```

O grafico mostra que a grande maioria das pessoas do titanic tinham em media 30 anos
